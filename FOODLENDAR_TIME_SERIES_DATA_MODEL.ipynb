```python
# 학습 파일 전처리 및 시각화
df = df[df['평균가격'] != '-']
df = df.reset_index(drop=True)
#df = df[df.index % 2 != 0]
df = df.reset_index(drop=True)
df
```

```python
df['평균가격'] = pd.to_numeric(df['평균가격'])

# 'DATE' 열을 datetime 타입으로 변환
df['DATE'] = pd.to_datetime(df['DATE'], format='%Y%m') 

# 연도와 월 추출
df['Year'] = df['DATE'].dt.year
df['Month'] = df['DATE'].dt.month

# 연도와 월을 기준으로 그룹화하고 평균 가격 계산
monthly_avg = df.groupby(['Year', 'Month'])['평균가격'].mean().reset_index()

# 시각화
plt.figure(figsize=(12, 6))
plt.plot(pd.to_datetime(monthly_avg[['Year', 'Month']].assign(day=1)), monthly_avg['평균가격'], marker='o', linestyle='-')
plt.title('Monthly Average Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from prophet import Prophet

train_df = df
# DATE 열을 datetime 타입으로 변환합니다.
train_df['DATE'] = pd.to_datetime(train_df['DATE'], format='%Y%m')

# Prophet 모델은 'ds'와 'y'라는 열 이름을 사용합니다.
train_df.rename(columns={'DATE': 'ds', '평균가격': 'y'}, inplace=True)
```

```python
# 모델 초기화 및 설정
# 모델은 교차검증과 그리드 서치를 사용합니다.
# 교차검증을 할때 MAE 기준으로 업데이트 됩니다.

from prophet import Prophet
from prophet.diagnostics import cross_validation, performance_metrics
from prophet.plot import plot_cross_validation_metric
from sklearn.model_selection import ParameterGrid
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


param_grid = {
    'changepoint_prior_scale': [0.01, 0.05, 0.1, 0.5, 0.8],
    'seasonality_prior_scale': [0.1, 1.0, 5, 10.],
    'seasonality_mode': ['multiplicative','additive' ]

}

# 최소 MAE를 저장할 변수
min_mae = float('inf')
best_params = {}
best_model = None


# 상한값과 하한값 설정 8%

cap = 1.08 * train_df['y'].max()  
floor = 0.92 * train_df['y'].min() 
train_df['cap'] = cap
train_df['floor'] = floor





# 파라미터 조합을 반복하며 모델 훈련 및 검증
for params in ParameterGrid(param_grid):
    model = Prophet(
        yearly_seasonality=True,
        daily_seasonality=False,
        seasonality_mode=params['seasonality_mode'],
        changepoint_prior_scale=params['changepoint_prior_scale'],
        seasonality_prior_scale=params['seasonality_prior_scale'],
        growth='logistic' 
    )

    model.fit(train_df) 
    df_cv = cross_validation(model, initial='1095 days', period='365 days', horizon='365 days')
    df_p = performance_metrics(df_cv)
    mae = df_p['mae'].mean()
    
    # MAE 업데이트
    if mae < min_mae:
        min_mae = mae
        best_params = params
        best_model = model

print(f"Best Parameters: {best_params}, with MAE: {min_mae}")

# 최적 모델을 사용하여 미래 데이터 예측
future = best_model.make_future_dataframe(periods=12, freq='M')
future['cap'] = cap
future['floor'] = floor
forecast = best_model.predict(future)
fig = best_model.plot(forecast)
plt.title('Forecast of Future Prices with Best Model')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.show()

# 최적화된 모델의 예측 결과 출력
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(12))

pre_ans = forecast['yhat'].tail(12)
```

```python
# 중요!!!!
# 앞에 셀에 있는 베스트 파라미터들을 따로 저장해 놨다가 나중에 2024 예측할때 사용해야 합니다.


df_pre_ans = pd.DataFrame(pre_ans)

df_pre_ans = df_pre_ans.reset_index(drop=True)
df_pre_ans.rename(columns={'yhat': '평균가격'}, inplace=True)
df_pre_ans.index = range(1, len(pre_ans) + 1)

df_pre_ans
```

```python
real_df = pd.read_excel('수박테스트2023.xlsx')
real_df = real_df[real_df['평균가격'] != '-']

real_df = real_df.reset_index(drop=True)
#real_df = real_df[real_df.index % 2 != 0]


avg_price = (real_df.loc[8, '평균가격'] + real_df.loc[9, '평균가격']) / 2

# 새로운 행 생성
new_row = {
    'DATE': 202303,
    '품목': '수박',
    '품종': '수박',
    '거래단위': '1개',
    '등급': '상품',
    '평균가격': avg_price,
    '전년': None, 
    '평년': None   
}

insert_index = real_df.index[real_df['DATE'] == 202304][0]

# 데이터프레임에 행 삽입
real_df = pd.concat([real_df.iloc[:insert_index], pd.DataFrame([new_row]), real_df.iloc[insert_index:]]).reset_index(drop=True)




real_df.drop(columns=['DATE', '품목', '품종'], inplace=True)
real_df = real_df.iloc[::-1]
real_df.index = range(1, len(pre_ans) + 1)


real_df
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
from sklearn.metrics import mean_absolute_error
import statsmodels.api as sm


# 예측값과 실제값 추출
y_true = real_df['평균가격']
y_pred = df_pre_ans['평균가격']


mae = mean_absolute_error(y_true, y_pred)
print(f"(MAE): {mae}")

# MSE 계산
mse = mean_squared_error(y_true, y_pred)
print(f"(MSE): {mse}")

# RMSE 계산
rmse = np.sqrt(mse)
print(f"(RMSE): {rmse}")


mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
print(f"(MAPE): {mape}%")

r2 = r2_score(y_true, y_pred)
print(f"(R²): {r2}")
```
